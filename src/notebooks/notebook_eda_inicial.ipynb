```python
#  Convierto la columna MatchID en el índice
df = df.set_index('MatchID')
```

```python
# Se filtra la temporada que se va a analizar y se guarda en otro dataframe
pl_23_24 = df[df.Season == "2023-2024"]
```

```python
# Para analizar el rendimiento local como visitante, y crear una tabla genérica con los datos de los equipos, se divide el dataframe anterior en dos:
# Uno con los resultados como local y otro como visitante de cada uno de los equipos.
resultados_en_casa = pl_23_24.groupby("HomeTeam")["FullTimeResult"].value_counts().unstack().rename(columns={"H": "Victorias", "D": "Empates", "A": "Derrotas"}).fillna(0).astype(int)
resultados_como_visitante = pl_23_24.groupby("AwayTeam")["FullTimeResult"].value_counts().unstack().rename(columns={"H": "Derrotas", "D": "Empates", "A": "Victorias"})
```

```python
# Se calculan los puntos obtenidos en función de los resultados, una columna muy importante para realizar los análisis
resultados_en_casa["Puntos"] = resultados_en_casa.Victorias * 3 + resultados_en_casa.Empates
resultados_como_visitante["Puntos"] = resultados_como_visitante.Victorias * 3 + resultados_como_visitante.Empates
```

```python
# Se traducen y se renombran las columnas para que todo el dataframe siga la misma estructura.
# Además, en el caso de resultados_como_visitante, se cambia el orden de las columnas y se ponen primero 
# las estadísticas del equipo visitante, ya que serían las propias de los equipos jugando fuera de casa.

nombres_columnas_local = {
    'FullTimeHomeTeamGoals': 'GolesAnotados',
    'FullTimeAwayTeamGoals': 'GolesRecibidos',
    'HalfTimeHomeTeamGoals': 'GolesDescansoAnotados',
    'HalfTimeAwayTeamGoals': 'GolesDescansoRecibidos',
    'HomeTeamShots': 'TirosRealizados',
    'AwayTeamShots': 'TirosRecibidos',
    'HomeTeamShotsOnTarget': 'TirosAPuerta',
    'AwayTeamShotsOnTarget': 'TirosRecibidosAPuerta',
    'HomeTeamCorners': 'CornersAFavor',
    'AwayTeamCorners': 'CornersEnContra',
    'HomeTeamFouls': 'FaltasCometidas',
    'AwayTeamFouls': 'FaltasRecibidas',
    'HomeTeamYellowCards': 'AmarillasPropias',
    'AwayTeamYellowCards': 'AmarillasRival',
    'HomeTeamRedCards': 'RojasPropias',
    'AwayTeamRedCards': 'RojasRival'
}

nombres_columnas_visitante = {
    'FullTimeAwayTeamGoals': 'GolesAnotados',
    'FullTimeHomeTeamGoals': 'GolesRecibidos',
    'HalfTimeAwayTeamGoals': 'GolesDescansoAnotados',
    'HalfTimeHomeTeamGoals': 'GolesDescansoRecibidos',
    'AwayTeamShots': 'TirosRealizados',
    'HomeTeamShots': 'TirosRecibidos',
    'AwayTeamShotsOnTarget': 'TirosAPuerta',
    'HomeTeamShotsOnTarget': 'TirosRecibidosAPuerta',
    'AwayTeamCorners': 'CornersAFavor',
    'HomeTeamCorners': 'CornersEnContra',
    'AwayTeamFouls': 'FaltasCometidas',
    'HomeTeamFouls': 'FaltasRecibidas',
    'AwayTeamYellowCards': 'AmarillasPropias',
    'HomeTeamYellowCards': 'AmarillasRival',
    'AwayTeamRedCards': 'RojasPropias',
    'HomeTeamRedCards': 'RojasRival'
}

nuevo_orden = [
    "Victorias", "Empates", "Derrotas", "Puntos",
    "GolesAnotados", "GolesRecibidos", "GolesDescansoAnotados", "GolesDescansoRecibidos",
    "TirosRealizados", "TirosRecibidos", "TirosAPuerta", "TirosRecibidosAPuerta",
    "CornersAFavor", "CornersEnContra", "FaltasCometidas", "FaltasRecibidas",
    "AmarillasPropias", "AmarillasRival", "RojasPropias", "RojasRival"
]

resultados_en_casa = resultados_en_casa.rename(columns=nombres_columnas_local)
resultados_como_visitante = resultados_como_visitante.rename(columns=nombres_columnas_visitante)[nuevo_orden]
```

```python
resultados_en_casa.sort_values("Puntos", ascending=False, inplace=True)
resultados_como_visitante.sort_values("Puntos", ascending=False, inplace=True)
```
